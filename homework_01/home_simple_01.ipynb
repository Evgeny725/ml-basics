```
# Задание 1.
#
# Пользователь в бесконечном цикле вводит числа через консоль до тех пор, пока не введёт что-то, отличное от числа (после этого цикл завершается). В конце программа должна вывести сумму всех введённых чисел.


number = True
sum_of_inputs = 0
# input_value = 0

def numeric_type_check(input_value):
    try:
        float(input_value)
        return True
    except (ValueError, TypeError):
        return False

while (number == True):
    input_value = input('enter any number to count OR any other symbol to finish')
    number = numeric_type_check(input_value)
    if number == True:
        sum_of_inputs = sum_of_inputs + float(input_value)
        print(f'{input_value}; current sum of inputs = {sum_of_inputs}')
    else:
        print('script stopped due to a non numeric input')
        break

```

```
# Задание 2.
#
# Пользователь вводит строку через консоль (состоящую только из букв в разном регистре). Программа должна создать новую строку, содержащую только те буквы из исходной строки, которые были в верхнем регистре.



def check_string_letters(input_value):
    return input_value.isalpha()

string_chars = False

while string_chars == False:
    input_value = input('enter letters only string (any case)')
    string_chars = check_string_letters(input_value)
    if check_string_letters(input_value) == False:
        print(f'>> The input string \n{input_value} \n>>contains forbidden char(s), please enter a-z A-Z only')
    else:
        print(f'\n>> original input string is: \n{input_value}')
        upper_case_result = str('')
        for char in input_value:
            if 'A' <= char <='Z':
                upper_case_result = upper_case_result + char
        print(f'>> Filtered upper case string is: \n{upper_case_result}  \n>> ({len(upper_case_result)} uppercase chars in total)')

```

```
# Задание 3.
#
# Классическая игра FizzBuzz. Программа в цикле выводит в консоль числа от 1 до 100.
# Если число кратно 3, вместо него выводится Fizz.
# Если кратно 5 — Buzz.
# Если кратно и 3, и 5 — FizzBuzz.


def check_divider(x):
    if x % 5 == 0 and x % 3 == 0:
        return "FizzBuzz"
    elif x % 5 == 0:
        return "Buzz"
    elif x % 3 == 0:
        return "Fizz"
    else:
        return ""

for x in range (1, 21):
    check_result = check_divider(x)
    print(f'x = {x} => {check_result}')
    x += 1

```

```
# Задание 4.
#
# Пользователь вводит строку, содержащую любые символы (буквы, цифры и специальные символы). Программа должна вычислить и вывести сумму всех цифр, присутствующих в этой строке.

input_string = input('enter your string with any symbols')

digits_from_input = []
x=0

for char in input_string:
    if char.isdigit():
       digits_from_input.append(char)
print(f'input string = {input_string}')
print(f'filtered digits: {digits_from_input}')
result_sum = sum(int(char) for char in digits_from_input)
print(f'total sum = {result_sum}')

```
