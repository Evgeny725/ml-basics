```python
file = open('data.txt', 'r', encoding='utf-8')
data = file.readlines()
file.close()
data = [item.strip().upper() for item in data]
print(data)

file = open('data.txt', 'a', encoding='utf-8')
file.write('\n'.join(data))
file.close()
file = open('data.txt', 'r', encoding='utf-8')
print(file.read())
```
