# I. Numpy

### Импортируйте NumPy

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
d1_array = np.zeros(10)
print(d1_array)
d1_array[4] = 1
print(d1_array)
d2_array = np.reshape(d1_array,(2,5))
print(d2_array)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
my_array = np.arange(start=10, stop=50, step=1)
print(my_array)

reversed_array = my_array[::-1]
print(f' reversed array:\n{my_array}')

even_array = reversed_array[reversed_array % 2 == 0]
print(f' even array:\n{even_array}')
```

```python
my_array = np.arange(0,9,1).reshape(3,3)
print(my_array)
```

# II. Pandas

### Импортируйте: pandas, matplotlib, seaborn

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

### Загрузите датасет Tips из набора датасетов seaborn

```python

my_dataset = sns.load_dataset("tips")
```

### Посмотрите на первые 5 строчек

```python
print(my_dataset.head())
```

### Узнайте сколько всего строчек и колонок в данных

```python
dataset_dim = my_dataset.shape
print(dataset_dim)
```

### Проверьте есть ли пропуски в данных

```python
print(my_dataset.isnull().sum())

print(my_dataset.isnull())

print(my_dataset[my_dataset.isnull().any(axis=1)])
```

### Посмотрите на распределение числовых признаков

```python
# print(my_dataset.value_counts())
print(my_dataset.head())

print(my_dataset.describe())

print(my_dataset.hist(bins=20, density=True, alpha=0.8, color='g', figsize=(10, 7)))
print(f' skewness: {my_dataset['total_bill'].skew()}')
```

### Найдите максимальное значение 'total_bill'

```python
print(my_dataset['total_bill'].max())
```

### Найдите количество курящих людей

```python
# ВАШ КОД ЗДЕСЬ
smoker_bills = len(my_dataset.query('smoker == "Yes"'))
print(f'total bills with smokers: {smoker_bills}')
# точное число курящих не известно, т.к. известен признак курения и общее число людей в чеке
smoker_size_sum = my_dataset.query('smoker == "Yes"')['size'].sum()
print(f'people in smoker bills: {smoker_size_sum}')
print(f'total people: {my_dataset['size'].sum()}')
```

### Узнайте какой средний 'total_bill' в зависимости от 'day'

```python

bill_mean_by_day = my_dataset.groupby('day')['total_bill'].mean()
print(bill_mean_by_day)
```

### Отберите строчки с 'total_bill' больше медианы и узнайте какой средний 'tip' в зависимости от 'sex'

```python

median_bill = my_dataset['total_bill'].median()
mean_bill = my_dataset['total_bill'].mean()
print(f"Median bill amount: ${median_bill:.3f}")
print(f"Mean bill amount: ${mean_bill:.3f}")
# медиана меньше среднего, что подтверждает перекос графика влево


greater_bills = my_dataset.query(f'total_bill > {median_bill}')
# print(greater_bills['total_bill'].sum())
tips_by_sex = greater_bills.groupby('sex')['tip'].mean()
print(tips_by_sex)
```

### Преобразуйте признак 'smoker' в бинарный (0-No, 1-Yes)

```python
my_dataset['smoker'] = my_dataset['smoker'].map({ 'No': 0, 'Yes': 1})
print(my_dataset)
```

### Постройте pairplot

```python
from pandas.plotting import scatter_matrix
scatter_matrix(my_dataset[['total_bill', 'tip']], grid=True,  s=30,
               alpha=0.8, figsize=(10, 8), diagonal='hist', color='cyan')

plt.suptitle('Total bill / Tips pairplot', y=0.9)
plt.show()
```

### Постройте график взаимосвязи между признаками 'total_bill' и 'day'

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
my_dataset = sns.load_dataset("tips")

print(my_dataset['day'].value_counts())

# box plot
plt.figure(figsize=(8, 3))
sns.boxplot(data = my_dataset, x = 'total_bill', y = 'day', color = 'cyan')
plt.title('Total bills, grouped by days')
plt.xlabel('Days')
plt.ylabel('Total bills')
plt.show()

# violin plot
plt.figure(figsize=(8, 3))
sns.violinplot(data=my_dataset, x='day', y='total_bill', color = 'cyan')
plt.xlabel('Days')
plt.ylabel('Total bills')
plt.show()

```


### Постройте две гистограммы распределения признака 'tip' в зависимости от категорий 'time'

```python
plt.figure(figsize=(12, 6))

# оказывается можно накладывать графики друг на друга
sns.boxplot(data = my_dataset, y ='time', x ='tip',
            color ='darkgreen', width = 0.3)


sns.swarmplot(data = my_dataset,  y ='time', x ='tip', hue = 'time',
              size = 5, legend=True, palette = 'Set3', alpha = 0.8)

plt.show()
```

### Постройте два графика scatterplot, представляющих взаимосвязь между признаками 'total_bill' и 'tip' один для Male, другой для Female и раскрасьте точки в зависимоти от признака 'smoker'

```python
from matplotlib import ticker

hist_smokers, axes = plt.subplots(1, 2, figsize = (12, 5))

colors = {'Yes': 'pink', 'No': 'cyan'}

for axis, gender in zip(axes, ['Male', 'Female']):

    gender_type = my_dataset[my_dataset['sex'] == gender]

    for smoker_status, group in gender_type.groupby('smoker'):
        axis.scatter(group['total_bill'], group['tip'],
                  c = colors[smoker_status],
                  label = f'Smoker: {smoker_status}',
                  alpha = 0.7, s = 30, edgecolors = 'black', linewidth = 0.5)

    axis.set_title(f'{gender}', fontsize=14, color = 'lightblue')
    axis.set_xlabel('Total bill')
    axis.set_ylabel('Tips')
    axis.set_ylim(0, 11)
    axis.set_xlim(0, 60)
    axis.legend()
    axis.grid(True, which = 'major', alpha = 0.5, linewidth=1)
    axis.grid(True, which = 'minor', alpha = 0.2, linewidth=0.8, linestyle=':')
    axis.xaxis.set_minor_locator(ticker.MultipleLocator(1))
    axis.yaxis.set_minor_locator(ticker.MultipleLocator(0.5))

    # amount
    axis.text(0.03, 0.96, f'amount = {len(gender_type)}',
            transform=axis.transAxes, fontsize=14,
            verticalalignment='top', bbox=dict(boxstyle='round', facecolor='gray', alpha=0.9))

plt.suptitle('Total bill and tips', fontsize=18, color = 'lightblue', y=1.0)
plt.show()
```
