### Импортируйте NumPy

```python
import numpy as np
```

### Создайте одномерный массив размера 10, заполненный нулями и пятым элемент равным 1. Трансформируйте в двумерный массив.

```python
d1_array = np.zeros(10)
print(d1_array)
d1_array[4] = 1
print(d1_array)
d2_array = np.reshape(d1_array,(2,5))
print(d2_array)
```

### Создайте одномерный массив со значениями от 10 до 49 и разверните его (первый элемент становится последним). Найдите в нем все четные элементы.

```python
my_array = np.arange(start=10, stop=50, step=1)
print(my_array)

reversed_array = my_array[::-1]
print(f' reversed array:\n{my_array}')

even_array = reversed_array[reversed_array % 2 == 0]
print(f' even array:\n{even_array}')
```

### Создайте двумерный массив 3x3 со значениями от 0 до 8

```python
my_array = np.arange(0,9,1).reshape(3,3)
print(my_array)
```

### Создайте массив 4x3x2 со случайными значениями. Найти его минимум и максимум.

```python
my_array = np.random.random((4,3,2))
# print(my_array)

print(np.min(my_array))
print(np.max(my_array))

```

### Создайте два двумерных массива размерами 6x4 и 4x3 и произведите их матричное умножение. 

```python
array_1 = np.random.random((6,4))
array_2 = np.random.random((4,3))

array_3 = array_1.dot(array_2)
print(np.shape(array_3))
print(array_3)
```

### Создайте случайный двумерный массив 7x7, найти у него среднее и стандартное оклонение. Нормализуйте этот массив.

```python
# my_array = np.random.random((7,7))
my_array = np.random.randint(0,100,(7,7))
print(my_array)

sredn = np.mean(my_array)
print(sredn)
std =  np.std(my_array)
print(std)
```
