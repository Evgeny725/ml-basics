```python
class Car:
    def start_engine(self):
        self.engine_on = True
    def drive_to(self, city):
        try:
            if self.engine_on:
                print('driving to', city)
        except Exception:
            print('engine is off')
    def stop_engine(self):
        self.engine_on = False

c = Car()
#c.start_engine()
c.drive_to('gorod')
```

```python
class Car:
    def __init__(self, color = 'red'): # init  чтобы создать экземпляр с такими то атрибутами по умолчанию
        self.engine_on = False
        self.color = color # цвет назначится при задании в агрументах при создании экземпляра
    def drive_to(self, city):
        try:
            if self.engine_on:
                print('driving to', city)
        except Exception:
            print('engine is off')

car1 = Car('green')
car1.color
```
