#### Даны матрицы:
$$
A = \begin{pmatrix}
-2 & -4 \\

-1 & 0
\end{pmatrix}, \quad
B = \begin{pmatrix}
1 & 2 & 7 \\
-3 & -4 & 0 \\
5 & 2 & 1
\end{pmatrix},
\quad
C = \begin{pmatrix}
6 & -3 & 9 \\
4 & -5 & 2 \\
8 & 1 & 5
\end{pmatrix}.
$$

Посчитать  $D = A^{T}C-2A^{T}B^{T}$

---
## Решение:
### 1. переворачиваем матрицы A и B
$$
A^{T} = \begin{pmatrix}
-2 & 3 & -1 \\
-4 & 5 & 0 \\
\end{pmatrix}, \quad
B^{T} = \begin{pmatrix}
1 & -3 & 5 \\
2 & -4 & 2 \\
7 & 0 & 1
\end{pmatrix}
$$

### 2. перемножаем  эти матрицы $A^{T}*B^{T}$
2х3 и 3х3 -> должна получиться матрица 2х3

берем первую строку из матрицы А, умножаем элементы на первый столбец В и все складываем = это элемент j11
$j11 = -2*1 + 3*2 + -1*7 = -3$

потом умножаем (и складываем) первую строку матрицы А на второй столбец матрицы В
$j12 = -2*-3 + 3*-4 + -1*0 = -6$

и первую строку матрицы А умножить на третий столбец матрицы В
$j13 = -2*5 + 3*2 + -1*1 = -5$

то же самое для второй строки из матрицы А, чтобы заполнить вторую строку в целевой матрице

$j21 = -4*1 + 5*2 + 0*7 = 6$

$j22 = -4*-3 + 5*-4 + 0*0 = -8$

$j23 = -4*5 + 5*2 + 0*1 = -10$

Итого $A^{T}*B^{T}$  = $\begin{pmatrix}-3 & -6 & -5 \\6 & -8 & -10 \\\end{pmatrix}$

### 3. умножаем матрицу на -2
каждый элемент матрицы умножаем на -2

Итого $-2*A^{T}*B^{T}$  = $\begin{pmatrix}6 & 12 & 10 \\-12 & 16 & 20 \\\end{pmatrix}$

### 4. перемножаем $A^{T}*С$

$$
A^{T} = \begin{pmatrix}
-2 & 3 & -1 \\
-4 & 5 & 0 \\
\end{pmatrix}\quad* \quad
C = \begin{pmatrix}
6 & -3 & 9 \\
4 & -5 & 2 \\
8 & 1 & 5
\end{pmatrix}
$$

первую строчку А на все столбцы С
вторую строчку А на все столбцы С

Итого $A^{T}*С$ = $\begin{pmatrix}-8 & -10 & -17 \\-4 & -13 & -26 \\\end{pmatrix}$


### 5. осталось сложить 2 матрицы
т.к. на -2 уже умножили выше ^

$\begin{pmatrix}6 & 12 & 10 \\-12 & 16 & 20 \\\end{pmatrix}$ + $\begin{pmatrix}-8 & -10 & -17 \\-4 & -13 & -26 \\\end{pmatrix}$

складываем по элементно

$\begin{pmatrix}6+-8 & 12+-10 & 10+-17 \\-12+-4 & 16+-13 & 20+-26 \\\end{pmatrix}$


#### Ответ:  D = $\begin{pmatrix}-2 & 2 & -7 \\-16 & 3 & -6 \\\end{pmatrix}$

# Задание 5
малоранговая аппроксимация матрицы

## Решение
---
### шаг 1
генерим случайную квадратную матрицу  $A \quad принадлежит \quad \mathbb{R}^2$:



```python
import numpy as np
n = 5 # строки и столбцы для матрицы

matrix_A = np.random.randint(-10,10,(n,n))
print(matrix_A)
```

---
### шаг 2
теперь надо умножить исходную матрицу на ее транспонированный вариант

```python
AtA = matrix_A.T @ matrix_A
print(AtA)
```

---
### шаг 3
теперь надо вычислить собственные векторы и собственные числа

```python
# Calculate eigenvalues (w) and eigenvectors (v)
eigenvalues, V_vectors = np.linalg.eigh(AtA)

# Sort them in descending order (SVD convention)
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
V_vectors = V_vectors[:, idx]
print(f'собственные значения \n{eigenvalues}')
print('-'*15)
print(f'собственные вектора \n{V_vectors}')
```

---
### шаг 4
дальше высчитываются квадратные корни от собственных значений, типа как коэф влияния или важности векторов. Чем больше сингулярное значение, тем важней этот вектор

```python
# Square root of eigenvalues = Singular Values
s = np.sqrt(np.maximum(eigenvalues, 0))
Sigma = np.diag(s)
print(f'сингулярные значения как корни из собственных чисел \nотсортированные от большего к меньшему \n{s}')
print('-'*15)
print(f' \n{Sigma}')
```

---
### шаг 5

построить матрицу **$V^H$** и матрицу **U**.
Где **$V^H$** описывает поворот к осям
**U** описывает как данные выглядят после трансформации

```python
# V-Hermitian is just the transpose of our V_vectors
Vh = V_vectors.T

# U is calculated by projected A onto V and normalizing by Sigma
# Or more simply, use the built-in function to get U directly:
U, s_internal, Vh_internal = np.linalg.svd(matrix_A)
print(f'Vh:\n{Vh_internal}')
print('-'*15)
print(f'U:\n{U}')
print('-'*15)
print(f's_internal:\n{s_internal}')
print('-'*15)
```

---
### Пример уменьшения этой матрицы до 5х2


```python
# Разложение на собств числа, векторы
U, s, Vh = np.linalg.svd(matrix_A)

# определяем ранг матрицы 2, чтобы получить 2 самые важные вектора и игнорировать остальные
k = 2

# 4. уменьшаем матрицу до размерности 5x2. т.е. новая матрица U 5х2 типа как новые координаты для исходной матрицы
# U страновится (5, 2), s -> (2,), Vh -> (2, 5)
U_k = U[:, :k]
s_k = s[:k]
Vh_k = Vh[:k, :]

# 5. восстанавливаем матрицу из сжатой матрицы, посмотреть насколько близко восстановится к исходным значениям

A_compressed = U_k @ np.diag(s_k) @ Vh_k

print(f'исходная матрица: \n{matrix_A}')
print('-'*15)
print(f'сжатая матрица: \n{A_compressed.round(3)}')
```

---
и проверка качества уменьшения матрицы


```python
total_energy = np.sum(s**2)
captured_energy = np.sum(s_k**2)
accuracy = (captured_energy / total_energy) * 100
print(f'точность: { accuracy:.3f}%')
```

### пример расчетов для размерности n >= 100

тут еще разобраться как это работает



```python
import numpy as np

n = 201
matrix_large_A = np.random.randint(-100,100,(n,n))

# 1. Get the singular values (this is the most "expensive" part)
U, s, Vh = np.linalg.svd(matrix_large_A)

# 2. Calculate the energy of each singular value
energies = s**2

# 3. Calculate the cumulative percentage of energy
# This creates an array: [energy1, energy1+2, energy1+2+3, ...]
cumulative_energy = np.cumsum(energies)
total_energy = np.sum(energies)
accuracy_levels = (cumulative_energy / total_energy) * 100

# 4. Find the first index where accuracy >= 95.5%
# np.argmax returns the first index where the condition is True
k = np.argmax(accuracy_levels > 95.5) + 1 # +1 потому что первый индекс нулевой

print(f"To reach 95.5% accuracy, use k = {k}")
print(f"Actual accuracy at k={k}: {accuracy_levels[k-1]:.2f}%")
```

# Задание 1

Посчитать градиент функции:
$f(x)=x_{1}^3 - 2x_{1}x_{2} + x_{2}^2 - 3x_{1} - 2x_{2}$

Найти критические точки $x_{c}$, такие что $\nabla f(x_{c})=0$


---

## Решение
---
### шаг 1

посчитать частные производные, чтобы понять скорость изменения по одной оси.
по x1:
$\frac{\partial f}{\partial x_1} = 3x_1^2 - 2x_2 - 3$

по x2: $\frac{\partial f}{\partial x_2} = -2x_1 + 2x_2 - 2$

получилась матрица, чтобы посчитать градиент:
$\nabla f(x) = \begin{pmatrix} 3x_1^2 - 2x_2 - 3 \\ -2x_1 + 2x_2 - 2 \end{pmatrix}$




---
### шаг 2

приравниваем уравнения к нулю, чтобы найти точку в которой производные равны нулю. т.е. когда векторы равны нулю, значит мы находимся в минимальной точке (а может в максимальной, не понял тут как отличить минимум от максимума, чат говорит, что надо считать вторую производную, если знак будет "+", то это ямка, если "-", то горка)

$\text{(уравнение 1)}\quad 3x_1^2 - 2x_2 - 3 = 0$

$\text{(уравнение 2)}\quad -2x_1 + 2x_2 - 2 = 0$

Решаем систему уравнений:

$-2x_1 + 2x_2 - 2 = 0 \quad \text{->} \quad 2x_2 = 2x_1 + 2 \quad \text{->} \quad x_2 = x_1 + 1$

Подставляем это в первое уравнение:
$3x_1^2 - 2(x_1 + 1) - 3 = 0\quad \text{->} \quad 3x_1^2 - 2x_1 - 5 = 0$

решаем квадратное уравнение:
  $x_1 = \frac{-b \pm \sqrt{(b)^2 - 4 \cdot a \cdot c}}{2 \cdot a}$


$x_1 = \frac{2 \pm \sqrt{(-2)^2 - 4 \cdot 3 \cdot (-5)}}{2 \cdot 3}$

$x_1 = \frac{2 \pm \sqrt{4 + 60}}{6}$

$x_1 = \frac{2 \pm 8}{6} \quad$  где $\quad x_1 = \frac{5}{3}$ или $\quad x_1 = -1$

  Находим x2:
  $x_2 = \frac{5}{3} + 1 = \frac{5}{3} + \frac{3}{3} = \frac{8}{3}\quad$
  или $\quad x_2 = -1 + 1 = 0$

---
### шаг 3
Проверяем градиент на 0 в полученных точках

$\mathbf{x}_c = \left(\frac{5}{3}, \frac{8}{3}\right) \quad \text{и} \quad \mathbf{x}_c =  (-1, 0)$

$\nabla f(x) = \begin{pmatrix} 3x_1^2 - 2x_2 - 3 \\ -2x_1 + 2x_2 - 2 \end{pmatrix}$

---

```python
# Method 2: Analytical solution using sympy
import sympy as sp

def gradient_equation():
    x1, x2 = sp.symbols('x1 x2')
    f = x1**3 - 2*x1*x2 + x2**2 - 3*x1 - 2*x2

    # градиент
    gradient = [sp.diff(f, x1), sp.diff(f, x2)]

    # приравнять к нулю
    critical_points = sp.solve(gradient, (x1, x2))

    return critical_points, gradient

if __name__ == "__main__":
    # символ частной производной \u2202
    points, gradient_values = gradient_equation()
    print("Решение")
    print(f"Gradient values: \u2202f/\u2202x1 = {gradient_values[0]}, \u2202f/\u2202x2 = {gradient_values[1]}")
    print(f"Critical points: {points}")
```
