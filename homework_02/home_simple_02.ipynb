```python
# Задание 1.
# Написать программу, которая получает на вход строку и возвращает словарь, где:
# ключи — символы из этой строки;
# значения — количество раз, сколько каждый символ встречается.

```

```python


input_string = input('enter a string to create a dictionary')
print(f'input string is {input_string}')
dict_from_string = dict

dict_from_string = {char: input_string.count(char) for char in input_string}



print(f' dict_from_string {type(dict_from_string)}')
print(f' result dictionary: \n{dict_from_string}')
```

```python
# Задание 2.
#
# Дана строка текста (или введённая через консоль). Программа должна вернуть новую строку, в которой порядок слов будет обратным.
# Пример:
# "Python is really cool" → "cool really is Python".

# constants
separators = {' ', '.', ','}
print(f'separators: {type(separators)}')

input_text = input('enter your text to reverse the words')
word = ''
word_list = []

print(f'input text: "{input_text}"')

for char in input_text:
    if char in separators:
        if word != '':
            word_list.append(str(word))
            word = ''
    else:
        word = word + char

# last word check:
if word != '':
    word_list.append(str(word))

print(f'input word list: {word_list} \n---')
print(f'reverse word list: {word_list[::-1]}')
```

```python
# Задание 4.
#
# Написать программу, которая удаляет из списка все дубликаты, сохранив исходный порядок элементов

test_list = ['qwe', 'wer', 'ert', 'qwe', 'rty', 'qwe', 'wer']

dict_test = dict.fromkeys(test_list)

print(test_list)
print(dict_test)

test_list = [key for key in dict_test.keys()]
print(f'{test_list} - no duplicates')
```

```python
# Задание 4.
#
# Даны три (или больше) списка с объектами. Программа должна создать новый список, содержащий все уникальные элементы — каждый объект встречается только один раз.

list_1 = ['q', 'w', 'e', 'q', 'r', 'q', 'w']
list_2 = ['a', 's', 'd', 'f', 'e', 'q', 'r']
list_3 = ['1', '2', '3', '4', '5', '6', '7']
list_4 = ['3', 'a', 'q', 'e', '1', 'c', 'b']

aggregated_dict = dict.fromkeys(list_1 + list_2 + list_3 + list_4)
print(aggregated_dict)

result_list = [key for key in aggregated_dict.keys()]
print(f'{result_list} - no duplicates')
```
