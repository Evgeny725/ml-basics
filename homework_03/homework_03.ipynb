## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
# ВАШ КОД ЗДЕСЬ



def check_char(char: str) -> str:
    if (char == '_'):
        return 'underscore'
    if ('A' <= char <= 'Z'):
        return 'uppercase'
    if ('a' <= char <= 'z' or
        '0' <= char <= '9'):
        return 'regular_char'
    else:
        return 'unrecognized_char'


def check_format(text: str) -> str:
    text = text.strip()
    text_length = len(text)
    #print(f'original text: {text}\n---')
    underscore_counter = 0
    upper_counter = 0
    regular_char_counter = 0
    unrecognized_counter = 0
    # if (lambda text: bool(text[0].isupper())):
    #     pascal_prob = 0.7
    # else:
    #      snake_prob = 0.5
    for char in text:
        if check_char(char) == 'underscore':
            underscore_counter +=1
        elif check_char(char) == 'uppercase':
            upper_counter +=1
        elif check_char(char) == 'regular_char':
            regular_char_counter +=1
        else:
            unrecognized_counter += 1
    #print(f'text length: {text_length}; where:\nunderscores: {underscore_counter}, upper cases: {upper_counter}, unrecognized chars: {unrecognized_counter} ')
    if (underscore_counter >= 1 and unrecognized_counter == 0 and upper_counter ==0):
        decision = 'snake_case'
        return decision
    elif (upper_counter >= 1 and unrecognized_counter == 0 and underscore_counter ==0):
        decision = 'pascal_case'
        return decision
    else:
        decision = 'unrecognized_format'
        return decision


def convert_format(text: str) -> str:
    text = text.strip()
    print(f'original text: {text}\n---')
    need_upper = False # flag to make uppercase
    converted_text = ''
    if check_format(text) == 'pascal_case': # make snake
        converted_text = text[0].lower()
        for i in range(1, len(text)-1):
            char = text[i]
            if char.isupper():
                converted_text = converted_text + '_' + str(char.lower())
            else:
                converted_text = converted_text + str(char)
        if text[len(text)-1].isupper():
            converted_text = converted_text + '_' + str(text[len(text)-1].lower())
        else:
            converted_text = converted_text + str(text[len(text)-1])
        return converted_text
    if check_format(text) == 'snake_case':
        converted_text = ''
        if text[0] != '_':
            converted_text = text[0].upper()

        for i in range(1, len(text)-1):
            char = text[i]
            if char == '_':
                need_upper = True
            else:
                if need_upper:
                    converted_text = converted_text + str(char).upper()
                    need_upper = False
                else:
                    converted_text = converted_text + str(char)
        converted_text = converted_text + str(text[len(text)-1])
        return converted_text
    if check_format(text) == 'unrecognized_format':
        return 'unrecognized_format'


input_text_1 = 'otus_course'
input_text_2 = 'PythonIsTheBest'
input_text_3 = 'textWith_error'

print(f'converted format is: "{convert_format(input_text_1)}"')
```

## Задание 2: Проверка валидности даты

Написать функцию проверяющую валидность введенной даты.

<br>

**Примеры:**
* `29.02.2000 -> True`
* `29.02.2001 -> False`
* `31.04.1962 -> False`



```python
# ВАШ КОД ЗДЕСЬ

def parse_date(input_date: str) -> tuple:
    day, month, year = map(int, input_date.split('.'))
    #print(type(day, month, year))
    return day, month, year


#syntax validations skipped


def check_date(parsed_date) -> bool:
    day = int(parsed_date[0])
    month = int(parsed_date[1])
    year = int(parsed_date[2])
    day_valid = False
    month_valid = False
    year_valid = False

    try:
        # check year is valid
        if 1000 <= year <= 9999:
            year_valid = True
        else:
            year_valid = False
        #check year is leap
        if (year % 100 == 0):
            if (year % 400 == 0):
                year_leap = True
            else:
                year_leap = False
        elif (year % 4 == 0):
            year_leap = True
        else: year_leap = False #to check - 1700 - false, 1988 - true, 2000 - true
        #check month is valid
        if 1 <= month <= 12:
            month_valid = True
        else:
            month_valid = False
        #check day is valid
        if day == 31:
            if month in (1, 3, 5, 7, 8, 10, 12):
                day_valid = True
        elif day == 30:
            if month in (4, 6, 9, 11):
                day_valid = True
        elif day == 29:
            if month == 2 and year_leap:
                day_valid = True
        elif 1 <= day <= 28:
            day_valid = True
        else:
            day_valid = False
        #return detailed result

        if day_valid and month_valid and year_valid:
           print(f'input date is valid')
           return True
        else:
           print(f'input date is invalid, details:\n')
           print(f'day valid:{day_valid}, month valid:{month_valid}, year_valid:{year_valid}, year_leap:{year_leap}')
    except:
           return False


input_date = '29.02.1988'
parsed_date = parse_date(input_date)
print(parsed_date)
check_date(parsed_date)

```

## Задание 3: Проверка на простое число

Функция проверки на простое число. Простые числа – это такие числа, которые делятся на себя и на единицу.

<br>

**Примеры:**
* `17 -> True`
* `20 -> False`
* `23 -> True`

```python
# ВАШ КОД ЗДЕСЬ
simple_list = [1, 2, 3, 5, 7, 11, 13, 17, 23]
simple_list_limit = 20

def check_number(number: int) -> bool:
    simple_list.sort()
    if number == 0:
        print(f'input number is 0, need number > 0')
        return None
    if number in simple_list:
        print(f'{number} in the simple_list')
        print(f'---\n{number} is simple ?')
        #check_result = True #is simple
        return True
    check_result = True
    print('')
    for i in range(1,(len(simple_list)-1)):
        divisor = simple_list[i]
        #print(f'check {number} / {divisor} = {round((number / divisor),2)}')
        if number % divisor == 0:
            check_result = False
            break

    if check_result == True and number > divisor:
        #print(f' checked number {number} > divisor {divisor} and we add it to simple_list ')
        if len(simple_list) < simple_list_limit:
            simple_list.append(number)
            print(f'list updated: {simple_list}')
        else:
            print('simple list length reached')

    print(f'---\n{number} is simple ?')
    return check_result

# print(check_number(17))
# print(check_number(20))
# print(check_number(29))
print(check_number(3499))
#
```

## Задание 4: Учет пользователей

Пользователь в бесконечном цикле вводит данные пользователей: имя, затем фамилию, возраст и ID. Ввод продолжается до тех пор, пока не будет введено пустое поле. 

Пользователи заносятся в словарь, где ключ это ID пользователя, а остальные данные записываются в виде кортежа. 

**Программа должна проверять:**
* имя и фамилия состоят только из символов и начинаются с большой буквы - если не с большой, то заменяет букву на большую;
* возраст должен быть числом от 18 до 60;
* ID - целое число, дополненное до 8 знаков незначащими нулями, ID должен быть уникальным.

**Дополнительно:** написать функцию, которая будет выводить полученный словарь в виде таблицы.

```python
# ВАШ КОД ЗДЕСЬ


def correct_name(input_text: str) -> str:
    name = ''
    for char in input_text:
        try:
            if char.isalpha():
                char = char.lower()
                if name == '':
                    name = char.upper()
                else:
                    name = name + char
                continue
            else:
                print('only letters a-z and A-Z allowed')
                return 'error'
        except Exception:
            return 'error'
    return name


def check_age(input_text: str) -> bool:
    try:
        if input_text.isdigit():
            if 18 <= int(input_text) <= 60:
                return True
            else:
                print('allowed age: 18-60')
                return False
        else:
            print('age should be an integer 18-60')
            return False
    except Exception:
        return False


def correct_id(input_text: str) -> str:
    try:
        if input_text.isdigit():
            id =  input_text.zfill(8)
            return id
        else:
            print('id should contain digits only')
    except Exception:
        print('error with id')
        return None


def remove_double_spaces(input_text: str) -> str:
    while '  ' in input_text:
        input_text = input_text.replace('  ', ' ')
    input_text = input_text.strip()
    return input_text


def add_user_data():
    input_text = str(' ')
    users_dict = {}
    while input_text != '':
        input_text = input(f'enter user name, surname, age, id. Separate values by space\n example: "john dow 18 1234"')
        input_text = remove_double_spaces(input_text)
        data_list = input_text.split(' ')
        #print(f'separated: {data_list}')
        if input_text == '':
            print('loop stopped')
            break
        if len(data_list) == 4:
            try:
                if correct_name(data_list[0]) != 'error':
                    data_list[0] = correct_name(data_list[0]) # name correction
                    print(f'name updated: {data_list[0]}')
                else:
                    continue
            except Exception:
                print(f'name error with "{data_list[0]}"')

            try:
                if correct_name(data_list[1]) != 'error':
                    data_list[1] = correct_name(data_list[1]) # surname correction
                    print(f'surname updated: {data_list[1]}')
                else:
                    continue
            except Exception:
                print(f'surname error with "{data_list[1]}"')

            if check_age(data_list[2]) == True: # check age range
                print(f'age "{data_list[2]}" ok')
            elif check_age(data_list[2]) == False:
                continue
            try:
                if correct_id(data_list[3]) == None:
                    continue
                elif correct_id(data_list[3]) == 0:
                    print('id must be greater "0"')
                    continue
                elif correct_id(data_list[3]) in users_dict.keys():
                    print(f"id '{data_list[3]}' already exists, please try another id")
                    continue
                data_list[3] = correct_id(data_list[3]) # id correction and check
            except Exception:
                print(f'id error with "{data_list[3]}"')

            dict_key = data_list[3]
            dict_value = tuple(data_list[0:3])
            users_dict.update({dict_key: dict_value})
            print(f"Dictionary updated: {users_dict}")
        elif 1 <= len(data_list) <= 999:
            print(f'there should be 4 items only, not {len(data_list)}')

        else:
            print(f'there should be 4 items only, not {len(data_list)}')
    return users_dict


def table_print(users_dict):
    print(f'\nuser data dictionary:')
    print(f'{"-"*50}')
    print(f'user_id  | name         | surname      | age')
    print(f'{"-"*50}')
    for key, value in users_dict.items():
        if len(value) == 3:
            name, surname, age = value
            print(f'{key:8} | {name:12} | {surname:12} | {age:2}')
        else:
            print(f'{key:8} | wrong number of attributes')
    print(f'{"-"*50}')


users_dict = add_user_data()
table_print(users_dict)

```
