## Задание 1: Конвертер регистров

Написать функцию, которая будет переводить snake_case в PascalCase и наоборот. 

Функция должна сама определять - какой формат ей передали. Можно добавить ключевой аргумент, который будет принудительно возвращать один из форматов.

<br>

**Примеры:**
* `otus_course     -> OtusCourse`
* `PythonIsTheBest -> python_is_the_best`



```python
# ВАШ КОД ЗДЕСЬ



def check_char(char: str) -> str:
    if (char == '_'):
        return 'underscore'
    if ('A' <= char <= 'Z'):
        return 'uppercase'
    if ('a' <= char <= 'z' or
        '0' <= char <= '9'):
        return 'regular_char'
    else:
        return 'unrecognized_char'



def check_format(text: str) -> str:
    text = text.strip()
    text_length = len(text)
    #print(f'original text: {text}\n---')
    underscore_counter = 0
    upper_counter = 0
    regular_char_counter = 0
    unrecognized_counter = 0

    # if (lambda text: bool(text[0].isupper())):
    #     pascal_prob = 0.7
    # else:
    #      snake_prob = 0.5

    for char in text:
        if check_char(char) == 'underscore':
            underscore_counter +=1
        elif check_char(char) == 'uppercase':
            upper_counter +=1
        elif check_char(char) == 'regular_char':
            regular_char_counter +=1
        else:
            unrecognized_counter += 1
    #print(f'text length: {text_length}; where:\nunderscores: {underscore_counter}, upper cases: {upper_counter}, unrecognized chars: {unrecognized_counter} ')
    if (underscore_counter >= 1 and unrecognized_counter == 0 and upper_counter ==0):
        decision = 'snake_case'
        return decision
    elif (upper_counter >= 1 and unrecognized_counter == 0 and underscore_counter ==0):
        decision = 'pascal_case'
        return decision
    else:
        decision = 'unrecognized_format'
        return decision
# ---

def convert_format(text: str) -> str:
    text = text.strip()
    print(f'original text: {text}\n---')
    need_upper = False # flag to make uppercase
    converted_text = ''
    if check_format(text) == 'pascal_case': # make snake
        converted_text = text[0].lower()
        for i in range(1, len(text)-1):
            char = text[i]
            if char.isupper():
                converted_text = converted_text + '_' + str(char.lower())
            else:
                converted_text = converted_text + str(char)
        if text[len(text)-1].isupper():
            converted_text = converted_text + '_' + str(text[len(text)-1].lower())
        else:
            converted_text = converted_text + str(text[len(text)-1])
        return converted_text
    if check_format(text) == 'snake_case':
        converted_text = ''
        for i in range(0, len(text)-1):
            char = text[i]
            if char == '_':
                need_upper = True
            else:
                if need_upper:
                    converted_text = converted_text + str(char).upper()
                    need_upper = False
                else:
                    converted_text = converted_text + str(char)
        converted_text = converted_text + str(text[len(text)-1])
        return converted_text
    if check_format(text) == 'unrecognized_format':
        return 'unrecognized_format'


input_text_1 = 'otus_course'
input_text_2 = 'PythonIsTheBest'
input_text_3 = 'textWith_error'

print(f'converted format is: "{convert_format(input_text_2)}"')
```
