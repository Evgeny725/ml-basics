```python
# Задание 1.
# Написать функцию, которая принимает неограниченное количество чисел в виде позиционных аргументов и ключевой аргумент — операцию над этими числами (сложение или умножение).
# Функция должна возвращать результат выполнения указанной операции.

def calculator(key_arg: str, *args, sep=','):
    if key_arg =='+':
        result_number = sum(args)
    elif key_arg =='*':
        result_number = 1
        for arg in args:
            result_number *= arg
    return result_number

print(calculator('*',1,2,3,4))

print(calculator('+',1,2,3,4))

```

```python
# Задание 2.
# Написать функцию для ввода из консоли целого числа.
#Если введено не число, функция должна вывести соответствующее сообщение и предложить ввести значение заново — до тех пор, пока пользователь не введёт корректное число.

def int_type_check(input_value):
    try:
        if int(input_value) ==True: # float to int check is skipped
            return True
    except (ValueError, TypeError):
        return False

input_value = input('enter integer: ')
while int_type_check(input_value) == False:
    input_value = input(f'the value {input_value} is not integer, enter integer: ')
print(f'input_value "{input_value}" is int')
```

```python
# Задание 3.
# Написать функцию, которая создаёт абсолютный путь к файлу.
# Позиционные аргументы:
# название диска,
# неограниченное количество папок,
# имя файла (без расширения).
#
# Ключевые аргументы:
# ext — расширение файла,
# sep — разделитель (по умолчанию '/').
#
# Пример:
# full_path('c:', 'work', 'python', 'function', 'main', ext='py') ➜ 'c:/work/python/function/main.py'


def full_path(disk: str, *folders: str, ext, sep='/'):
    folder_path = [disk] + list(folders)
    #print(folder_path)
    full_path = f'{sep.join(folder_path)}.{ext}'
    #print(full_path)
    return full_path

print(full_path('c:', 'work', 'python','folder1', 'folder2', 'function', 'main', ext='py'))

```

```python
# Задание 4.
# Написать функцию, которая принимает список, состоящий из объектов разных типов, и возвращает словарь, где:
# ключи — типы данных объектов;
# значения — списки объектов соответствующего типа.

list_1 = [1,[1,2,3],'abc',3.4, {'a', 'b', 'c'}]
print(list_1)

def object_type_dict(list_1:list):
    result_dict = {}
    for item in list_1:
        key = type(item)
        result_dict[key] = item
    return result_dict

print(object_type_dict(list_1))
# same types replaced with the last one from the list_1
```
